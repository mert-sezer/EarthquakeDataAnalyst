# Kİ KARE DOĞRULUK TESTİ DEVAMI

```python
import pandas as pd
import itertools
from scipy.stats import chi2_contingency

# Veriyi YÜKLENİYOR.
data = pd.read_csv('deprem_senaryosu.csv')

# KATEGORİK DEĞİŞKENLER BELİRLENİYOR.
categorical_columns = data.columns

# İKİLİ KARŞILAŞTIRMA YAPILIYOR.
combinations = list(itertools.combinations(categorical_columns, 2))

# SONUÇLARI DEPOLAMAK İÇİN RESULT KOMUTU KULLANIYORUZ.
results = []

# Tüm kombinasyonlar için ki-kare testi yapalım.
for combo in combinations:
    table = pd.crosstab(data[combo[0]], data[combo[1]])
    chi2, p, dof, ex = chi2_contingency(table)
    results.append({
        'Variable 1': combo[0],
        'Variable 2': combo[1],
        'Chi2': chi2,
        'p-value': p,
        'Degrees of Freedom': dof
    })

# Sonuçları bir DataFrame olarak gösterelim
results_df = pd.DataFrame(results)

# Anlamlı sonuçları filtreleyelim (örneğin p-value < 0.05)
significant_results = results_df[results_df['p-value'] < 0.05]

# Sonuçları yazdıralım
print(significant_results)
```
